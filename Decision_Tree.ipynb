```python
df.drop(['Name'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['Sex']=lb.fit_transform(df['Sex'])
```

```python
df['Ticket']=lb.fit_transform(df['Ticket'])
df['Cabin']=lb.fit_transform(df['Cabin'])
df['Embarked']=lb.fit_transform(df['Embarked'])
```

```python
df=df.drop(['PassengerId','Ticket','Cabin'],axis=1)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='median')
```

```python
df['Age']=imputer.fit_transform(df[['Age']])
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
x=df.iloc[:,1:]
```

```python
x.columns
```

```python
y=df.iloc[:,0]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_score=cross_val_score(dt,x,y,cv=5)
```

```python
cv_score
```

```python
import numpy as np
```

```python
np.mean(cv_score)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
cv_score1=cross_val_score(lr,x,y,cv=5)
```

```python
np.mean(cv_score1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=123)
```

```python
dt_final=DecisionTreeClassifier()
```

```python
dt_final.fit(xtrain,ytrain)
```

```python
pred=dt_final.predict(xtest)
```

```python
lr_final=LogisticRegression()
```

```python
lr_final.fit(xtrain,ytrain)
```

```python
pred_lr=lr_final.predict(xtest)
```

```python
df['Survived'].value_counts()
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(ytest,pred),accuracy_score(ytest,pred_lr)
```

```python
confusion_matrix(ytest,pred),confusion_matrix(ytest,pred_lr)
```

```python
print(classification_report(ytest,pred),classification_report(ytest,pred_lr))
```
